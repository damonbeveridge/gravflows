```python
import numpy as np
import matplotlib.pyplot as plt

# series = 'freqseries'
# series = 'timeseries'
series = 'test_snr_gen'

test_waveforms = np.load("./data/"+series+"/waveforms.npy")
test_projections = np.load("./data/"+series+"/projections.npy")
test_noise = np.load("./data/"+series+"/noise.npy")
test_snrs = np.load("./data/"+series+"/snrs.npy")
```

```python
print(test_waveforms.shape)
print(test_projections.shape)
print(test_noise.shape)
print(test_snrs.shape)
```

```python
# For TimeSeries
# Shape is formatted as below:
# (sample_idx, ifo, sample)
# ie. (0, H1, 0s...9s)
#     (0, L1, 0s...9s)
#     .
#     .
#     .
#     (n, H1, 0s...9s)
#     (n, L1, 0s...9s)
# For simplicity, I have used the number of seconds instead of number of datapoints in the timeseries
```

```python
# For FrequencySeries
# Shape is formatted as below:
# (sample_idx, ifo, sample)
# ie. (0, H1, 0...8192)
#     (0, L1, 0...8192)
#     .
#     .
#     .
#     (n, H1, 0...8192)
#     (n, L1, 0...8192)
# Here, the numbers in the third slot refer to the bin of the frequency series (which has 8193 total bins in this case)
```
