```python
import numpy as np
from numpy.lib.format import open_memmap

memmap = open_memmap('test.npy', mode='w+', dtype=np.float64, shape=(10000, 10, 2, 4096))

memmap
```

```python
import time

start_time = time.time()

num_elements = memmap.size
num_bytes = num_elements * 8 + 128 # 64 bit floats so 8 bytes for element
num_kbytes = num_bytes / 1024
num_mbytes = num_kbytes / 1024
num_gbytes = num_mbytes / 1024
num_tbytes = num_gbytes / 1024

end_time = time.time()
print("Time taken: " + str(end_time-start_time) + " seconds")

print(num_bytes)
print(num_kbytes)
print(num_mbytes)
print(num_gbytes)
print(num_tbytes)
```

```python
import os
import time

start_time = time.time()

x = os.path.getsize('test.npy') # returns in bytes
kb_x = x / 1024
mb_x = kb_x / 1024
gb_x = mb_x / 1024
tb_x = gb_x / 1024

end_time = time.time()
print("Time taken: " + str(end_time-start_time) + " seconds")

print(x)
print(kb_x)
print(mb_x)
print(gb_x)
print(tb_x)
```
