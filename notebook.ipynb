```python
import numpy as np
import matplotlib.pyplot as plt
from pycbc.types.frequencyseries import FrequencySeries
from pycbc.types.timeseries import TimeSeries

train_waveforms = np.load("./data/train/waveforms.npy")

val_waveforms = np.load("./data/validation/waveforms.npy")
val_projections = np.load("./data/validation/projections.npy")

test_waveforms = np.load("./data/test/waveforms.npy")
test_projections = np.load("./data/test/projections.npy")
test_noise = np.load("./data/test/noise.npy")
```

```python
print(train_waveforms.shape)

print(val_waveforms.shape)
print(val_projections.shape)

print(test_waveforms.shape)
print(test_projections.shape)
print(test_noise.shape)
```

```python
sample = 1
train_waveforms_h1_freq = FrequencySeries(test_waveforms[sample][0], delta_f=1)
train_waveforms_l1_freq = FrequencySeries(test_waveforms[sample][1], delta_f=1)

plt.plot(abs(train_waveforms_h1_freq))
plt.plot(abs(train_waveforms_l1_freq))
```

```python
train_waveforms_h1_time = train_waveforms_h1_freq.to_timeseries()
train_waveforms_l1_time = train_waveforms_l1_freq.to_timeseries()

plt.plot(train_waveforms_h1_time)
plt.plot(train_waveforms_l1_time)
```

```python
train_waveforms_h1_freq = FrequencySeries(test_projections[sample][0], delta_f=1)
train_waveforms_l1_freq = FrequencySeries(test_projections[sample][1], delta_f=1)

plt.plot(abs(train_waveforms_h1_freq))
plt.plot(abs(train_waveforms_l1_freq))
```

```python
train_waveforms_h1_time = train_waveforms_h1_freq.to_timeseries()
train_waveforms_l1_time = train_waveforms_l1_freq.to_timeseries()

plt.plot(train_waveforms_h1_time)
plt.plot(train_waveforms_l1_time)
```
